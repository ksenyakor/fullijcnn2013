```python
from PIL import Image
import random
import numpy as np
from __future__ import print_function
import os, sys
from object_detection.utils import dataset_util
from object_detection.utils import label_map_util

import tensorflow as tf
```

```python
output_name = 'data/FullIJCNN_label_map.pbtxt'
input_label_file = 'data/Labels.txt'
with open(input_label_file) as f:
    content = f.readlines()
content = [x.strip() for x in content] 
```

```python
end = '\n'
s = ' '
for x in content:

    row=x.split('=')
    row = [z.strip() for z in row] 
    ID = int(row[0])
   
    names = row[1].split('(')
    names = [z.strip() for z in names] 
    name = names[0]
    out = ''
    out += 'item' + s + '{' + end
    out += s*2 
    out += 'id:' + ' ' 
    out += (str(ID+1)) + end
    out += s*2 + 'name:' + ' ' + '\'' + name + '\'' + end
    out += '}' + end*2
    

    with open(output_name, 'a') as f:
        f.write(out)
```
