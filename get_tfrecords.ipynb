```python
from PIL import Image
import random
import numpy as np
from __future__ import print_function
import os, sys
import zipfile
import urllib.request
import shutil
from object_detection.utils import dataset_util
from object_detection.utils import label_map_util

import tensorflow as tf
```

```python
url = 'http://benchmark.ini.rub.de/Dataset_GTSDB/FullIJCNN2013.zip'
zip_path = 'data/FullIJCNN2013.zip'
with urllib.request.urlopen(url) as response, open(zip_path, 'wb') as out_file:
    shutil.copyfileobj(response, out_file)
```

```python
with zipfile.ZipFile(zip_path,"r") as zip_ref:
    zip_ref.extractall('data/')
```

```python
output_path = 'data/FullIJCNN2013/'
input_gt_file = 'data/FullIJCNN2013/gt.txt'
with open(input_gt_file) as f:
    content = f.readlines()
content = [x.strip() for x in content] 
```

```python
#creating and/or reading jpeg images to images list
images=[]

for x in content:
    image={}
    row=x.split(';')
    infile= 'data/FullIJCNN2013/'+row[0]
    f, e = os.path.splitext(infile)
    outfile = f + ".jpg"
    try:
        if not os.path.isfile(outfile): 
            im = Image.open(infile)
            im.save(outfile)
            print(outfile+' saved')
            os.remove(infile)
            
            
        matches = [match_im for match_im in images if match_im['filename'] == outfile ]
            
        if  len(matches)==0:
            xmin = []
            xmax = []
            ymin = []
            ymax = []
            id_class=[]
            #print(infile, im.format, im.size, im.mode)
            im2 = Image.open(outfile)
            image['filename']=outfile
            image['width']=im2.width
            image['height']=im2.height
            xmin.append(int(row[1])/im2.width)
            image['xmin']=xmin
            ymin.append(int(row[2])/im2.height)
            image['ymin']=ymin
            xmax.append(int(row[3])/im2.width)
            image['xmax']=xmax
            ymax.append(int(row[4])/im2.height)
            image['ymax']=ymax
            id_class.append(int(row[5]))
            image['id']=id_class
            images.append(image)
        else:
            ind = images.index(matches[0])
            image = images[ind]
            #print(images.index(matches[0]))
            width=image['width']
            height= image['height']
            xmin = image['xmin']
            xmax = image['xmax']
            ymin = image['ymin']
            ymax = image['ymax']
            id_class = image['id']
            xmin.append(int(row[1])/width)
            image['xmin']=xmin
            ymin.append(int(row[2])/height)
            image['ymin']=ymin
            xmax.append(int(row[3])/width)
            image['xmax']=xmax
            ymax.append(int(row[4])/height)
            image['ymax']=ymax
            id_class.append(int(row[5]))
            image['id']=id_class
            images[ind] = image
           
    except IOError:
        print("cannot convert", infile)
```

```python
n= len(images)
```

```python
mask = random.sample(range(n), int(n*0.7)) #mask for creating training dataset
```

```python
def create_tf_example(example):
    height = example['height'] # Image height
    width = example['width'] # Image width
    
    filename = example['filename'] # Filename of the image. Empty if image is not from file
    with tf.gfile.GFile(filename, 'rb') as fid:
        encoded_image_data = fid.read() # Encoded image bytes
        #print(len(encoded_image_data))
    image_format = b'jpeg' # b'jpeg' or b'png'
    #print(filename)
    xmins = [x for x in example['xmin'] ]# List of normalized left x coordinates in bounding box (1 per box)
    xmaxs = [x for x in example['xmax'] ]  # List of normalized right x coordinates in bounding box
             # (1 per box)
    ymins = [y for y in example['ymin'] ]  # List of normalized top y coordinates in bounding box (1 per box)
    ymaxs = [y for y in example['ymax'] ] # List of normalized bottom y coordinates in bounding box
             # (1 per box)
    classes_text = [str(i).encode('utf8') for i in example['id']] # List of string class name of bounding box (1 per box)
    classes = example['id'] # List of integer class id of bounding box (1 per box)

    tf_example = tf.train.Example(features=tf.train.Features(feature={
      'image/height': dataset_util.int64_feature(height),
      'image/width': dataset_util.int64_feature(width),
      'image/filename': dataset_util.bytes_feature(filename.encode('utf8')),
      'image/source_id': dataset_util.bytes_feature(filename.encode('utf8')),
      'image/encoded': dataset_util.bytes_feature(encoded_image_data),
      'image/format': dataset_util.bytes_feature(image_format),
      'image/object/bbox/xmin': dataset_util.float_list_feature(xmins),
      'image/object/bbox/xmax': dataset_util.float_list_feature(xmaxs),
      'image/object/bbox/ymin': dataset_util.float_list_feature(ymins),
      'image/object/bbox/ymax': dataset_util.float_list_feature(ymaxs),
      'image/object/class/text': dataset_util.bytes_list_feature(classes_text),
      'image/object/class/label': dataset_util.int64_list_feature(classes),
    }))
    return tf_example
```

```python
writer = tf.python_io.TFRecordWriter('data/training.record')
writer_test = tf.python_io.TFRecordWriter('data/testing.record')

for i in range(n):
    example= images[i]
    tf_example = create_tf_example(example)
    if i in mask:
        writer.write(tf_example.SerializeToString())
    else:
        writer_test.write(tf_example.SerializeToString())

writer.close()
writer_test.close()
```
